```python
import pandas as pd
```

```python
path = '../../../../src/main/resources/SimpleRegressionDataset/simple_regression.csv'
```

```python
data = pd.read_csv(path)
```

```python
data.head(n=10)
```

```python
data.columns=['population','profit']
```

```python
data.sort_values(['population'], inplace=True)
```

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
pop = data['population']
prof = data['profit']
fig = plt.plot(pop, prof, lw=2)
plt.title('Training Data')
plt.xlabel('population')
plt.ylabel('profit')
plt.show()
```

```python
pred_path= '../../../../src/main/resources/SimpleRegressionDataset/predictions.csv'
```

```python
data1 = pd.read_csv(path)
data1.columns=['population','profit']
```

```python
data2 = pd.read_csv(pred_path)
data2.columns=['prediction']
data2.head(10)
```

```python
#data2 = pd.read_csv(pred_data)
raw_data = {
        'population': data1['population'],
        'profit': data1['profit'], 
        'prediction': data2['prediction']}
df_a = pd.DataFrame(raw_data, columns = ['population', 'profit','prediction'])
df_a

```

```python
df_a.sort_values(['population'], inplace=True)
pop = df_a['population']
prof = df_a['profit']
prediction = df_a['prediction']
plt.plot(pop, prof, lw=2,label="profit")
plt.plot(pop, prediction, lw=2,label="predicted_profit")
plt.title('Actual vs Predicted Profit (using TransmogrifAI)')
plt.xlabel('population')
plt.ylabel('')
plt.legend()
plt.show()
```
